#### Q1: Remove Duplicates from List using Set

```python
my_list = [1, 2, 2, 3, 4, 4, 5, 1]
unique_list = list(set(my_list))
print(unique_list)
```

#### Q2: Print Dictionary Keys

#### Q3: Print Dictionary Keys & Values

```python
student_scores = {"daniyal": 90, "usman": 89, "ali": 67}

for name, score in student_scores.items():
    print(f"Name: {name} | Score: {score}")
```

#### Q4: Numbers divisible by 3 and 5 (Range 1-20)

```python
result = [x for x in range(1, 21) if x % 3 == 0 and x % 5 == 0]

print(result)
```

#### Q10: Count Vowels in String

```python
text = input("Enter a string: ")
vowels = "aeiouAEIOU"

count = 0
for ch in text:
    if ch in vowels:
        count += 1

print("Number of vowels:", count)
```

#### Q11: Reverse List using Loop

```python
my_list = [1, 2, 3, 4, 5]
reversed_list = []

for i in range(len(my_list) - 1, -1, -1):
    reversed_list.append(my_list[i])

print(reversed_list)
```

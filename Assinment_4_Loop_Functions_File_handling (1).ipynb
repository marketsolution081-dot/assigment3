**Q12: Check karein ke number Prime hai ya nahi.**

```python
### Apne code yahan likhein
num = int(input("Number enter karein: "))

is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("Prime number hai")
else:
    print("Prime number nahi hai")
```

#### **Q20: Number Guessing Game (3 chances).**

```python
# Apne code yahan likhein
import random

def guess_the_number():
    secret_number = random.randint(1, 10)
    attempts = 3
    
    print("--- Welcome to the Number Guessing Game! ---")
    print("I'm thinking of a number between 1 and 10. You have 3 chances.")

    for i in range(1, attempts + 1):
        try:
            guess = int(input(f"Attempt {i}: Enter your guess: "))
            
            if guess == secret_number:
                print(f"🎉 Correct! The number was {secret_number}. You win!")
                return
            elif guess < secret_number:
                print("Too low!")
            else:
                print("Too high!")
        except ValueError:
            print("Please enter a valid whole number.")

    print(f"❌ Game over! The secret number was {secret_number}.")

guess_the_number()
```

**Q23: Arbitrary arguments (`*args`) ka use karke function banayein jo saare numbers ko multiply kare.**

```python
# Apne code yahan likhein
def multiply_numbers(*args):
    result = 1
    
    for num in args:
        result *= num
        
    return result


print(multiply_numbers(2, 3, 4))
```

**Q24: Keyword arguments (`**kwargs`) ka use karke student ki info print karein (Name, Age).**

```python
# Apne code yahan likhein
def student_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

student_info(Name="Ali", Age=20)
```

#### **Q29: Recursive function banayein jo factorial calculate kare.**

```python
# Apne code yahan likhein
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Function call
print(factorial(5))
```

#### **Q30: Function banayein jo user se unlimited numbers le (*args) aur sirf even numbers ki list return kare.**

```python
# Apne code yahan likhein
def get_even_numbers(*args):
    even_numbers = []

    for num in args:
        if num % 2 == 0:
            even_numbers.append(num)

    return even_numbers

# Function call
print(get_even_numbers(1, 2, 3, 4, 5, 6, 7, 8))
```
